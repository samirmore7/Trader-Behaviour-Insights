```python
median_leverage = merged['leverage_proxy'].median()

merged['leverage_group'] = np.where(
    merged['leverage_proxy'] >= median_leverage,
    'High Leverage',
    'Low Leverage'
)
```

```python
trade_counts = merged['Account'].value_counts()

median_trades = trade_counts.median()

frequent_traders = trade_counts[trade_counts >= median_trades].index

merged['frequency_group'] = merged['Account'].apply(
    lambda x: 'Frequent' if x in frequent_traders else 'Infrequent'
)
```

```python
win_rate_trader = merged.groupby('Account')['win'].mean()

median_win_rate = win_rate_trader.median()

consistent_traders = win_rate_trader[win_rate_trader >= median_win_rate].index

merged['consistency_group'] = merged['Account'].apply(
    lambda x: 'Consistent' if x in consistent_traders else 'Inconsistent'
)
```

Methodology

The objective of this analysis was to understand how market sentiment influences trader behavior and performance.

First, both datasets were cleaned and validated by checking for missing values and duplicates. Timestamps were standardized and aligned at a daily level to ensure accurate merging of trading data with sentiment data.

After merging the datasets, key metrics were created such as daily PnL, win rate, trade size, and a leverage proxy based on position size. The analysis was then conducted in three stages:

Comparing trader performance across different sentiment conditions

Studying how trader behavior changes with sentiment (trade size, leverage, frequency)

Segmenting traders based on leverage usage, trading frequency, and consistency of wins

Visualizations and summary statistics were used to support all findings.

Key Insights

The analysis shows a clear relationship between market sentiment and trader performance.

Traders tend to perform better during bullish sentiment periods, especially during Extreme Greed, where both average PnL and win rate are highest. In contrast, performance declines during Extreme Fear, where win rates are lowest.

Trader behavior also changes significantly depending on sentiment. During Fear conditions, traders take larger positions and use higher leverage, indicating more aggressive risk-taking. However, this does not always translate into higher win rates.

Segmentation analysis revealed important behavioral patterns. High-leverage traders generate significantly higher profits compared to low-leverage traders. Infrequent traders tend to outperform frequent traders, suggesting that selective trading decisions may be more effective than overtrading. Interestingly, inconsistent traders show higher average PnL, likely due to high-risk, high-reward strategies.

These findings highlight that both sentiment and trading style play a major role in profitability.

Strategy Recommendations

Based on the findings, the following strategic recommendations can be considered:

1. Risk control during Fear sentiment

Since traders increase leverage and trade size during Fear periods, risk exposure becomes higher. Implementing leverage limits and position size controls during such periods can help reduce potential drawdowns and protect capital.

2. Encourage selective trading behavior

Infrequent traders and high-leverage traders tend to generate higher returns. Instead of promoting frequent trading, platforms can encourage high-conviction trades supported by strong signals and disciplined entry strategies.

3. Behavior-based trader segmentation

Risk management strategies should be customized based on trader type. High-risk traders may require tighter controls, while consistently profitable traders can be given more flexibility in leverage and position sizing.

Conclusion

This analysis demonstrates that market sentiment has a measurable impact on both trader performance and behavior. Traders become more aggressive during Fear conditions and more profitable during Greed-driven markets.

Understanding these behavioral patterns can help trading platforms design better risk management systems, improve trader guidance, and optimize overall performance outcomes.
